#Theory Questions

1. What is a random variable in probability theory?
*  A random variable is a variable that takes on different numerical values based on the outcomes of a random experiment. It's like a function that assigns numbers to each possible outcome. For example, when tossing a coin, we can define a random variable where Heads = 1 and Tails = 0.



2. What are the types of random variables?
*  There are mainly two types of random variables: discrete and continuous. A discrete random variable takes specific, countable values (like the number of heads in 3 coin tosses), while a continuous random variable can take any value within a range (like the height of students in a class).



3. What is the difference between discrete and continuous distributions?
*  The main difference is that discrete distributions deal with countable outcomes, like the number of emails we get in a day, while continuous distributions deal with measurable outcomes, like the time we spend waiting for a bus. Discrete values are separate, but continuous values form a range.



4. What are probability distribution functions (PDF)?
*  Probability Distribution Functions (PDFs) describe how probabilities are distributed over the values of a random variable. For a discrete variable, we use a probability mass function (PMF), and for a continuous variable, we use a probability density function (PDF). PDFs help us understand the likelihood of different outcomes.



5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
*  The PDF tells us the likelihood of a specific value (for discrete) or a small range (for continuous), while the CDF gives the probability that a random variable is less than or equal to a certain value. In simple terms, PDF shows "how likely," and CDF shows "how much so far."



6. What is a discrete uniform distribution?
*  A discrete uniform distribution is one where all outcomes have an equal chance of occurring. For example, when we roll a fair six-sided die, each number from 1 to 6 has a probability of 1/6. The probabilities are evenly spread across the possible values.



7. What are the key properties of a Bernoulli distribution?
*  A Bernoulli distribution models a random experiment with only two outcomes: success (1) and failure (0). Its key properties are:

  *  The probability of success is p, and failure is 1 - p.
  *  The mean is p, and the variance is p(1 - p).
It’s the simplest form of a discrete distribution.



8. What is the binomial distribution, and how is it used in probability?
*  The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has two outcomes (success or failure) and the same probability of success. We use it when we're counting how many times an event happens, like getting 3 heads in 5 coin tosses.



9. What is the Poisson distribution and where is it applied?
*  The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, when the events happen independently and at a constant average rate. We often use it to predict things like the number of customer calls in an hour or traffic accidents in a day.



10. What is a continuous uniform distribution?
*  A continuous uniform distribution is where every value within a given range is equally likely to occur. For example, if we pick a number randomly between 2 and 10, each number in that range has the same chance. The probability is spread evenly over the interval.



11. What are the characteristics of a normal distribution
*  A normal distribution is symmetric and bell-shaped, centered around the mean. Its key characteristics are:
    *  The mean, median, and mode are all equal.
    *  Data near the mean are more frequent, and probability decreases as we move away.
    *  It’s defined by two parameters: mean  and standard deviation.



12. What is the standard normal distribution, and why is it important?
*  The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It's important because it allows us to standardize data using z-scores, making it easier to compare values from different normal distributions and use statistical tables.



13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
*  The Central Limit Theorem (CLT) states that if we take large enough random samples from any population, the sampling distribution of the sample mean will be approximately normal, regardless of the population's original distribution. It's critical because it lets us use normal distribution techniques for inference, even when data isn’t normally distributed.



14. How does the Central Limit Theorem relate to the normal distribution?
*  The Central Limit Theorem explains why the normal distribution appears so often in statistics. It tells us that as we increase the sample size, the distribution of the sample mean becomes approximately normal, even if the original data isn't. This makes the normal distribution a powerful tool for making predictions and confidence intervals.



15. What is the application of Z statistics in hypothesis testing?
*  Z statistics are used in hypothesis testing when the population standard deviation is known and the sample size is large. We calculate the z-score to measure how far the sample statistic is from the population parameter under the null hypothesis. It helps us decide whether to reject or accept the null hypothesis based on standard normal distribution.



16. How do you calculate a Z-score, and what does it represent?
*  We calculate a Z-score using the formula:
Z = (X - μ) / σ,
where X is the data point, μ is the mean, and σ is the standard deviation. It tells us how many standard deviations a value is from the mean, helping us understand whether it's typical or unusual.



17. What are point estimates and interval estimates in statistics?
*  A point estimate gives us a single best guess for a population parameter, like using the sample mean to estimate the population mean. An interval estimate gives a range of values, like a confidence interval, which tells us where the true value is likely to fall with a certain level of confidence.



18. What is the significance of confidence intervals in statistical analysis?
*  Confidence intervals give us a range within which we expect the true population parameter to lie, based on sample data. They show not just the estimate but also the degree of uncertainty around it. This helps us make more informed decisions rather than relying on a single point estimate.



19. What is the relationship between a Z-score and a confidence interval?
*  The Z-score helps determine the margin of error for a confidence interval when the population standard deviation is known. For example, a Z-score of 1.96 corresponds to a 95% confidence level. So, we use Z-scores to calculate how wide the confidence interval should be around a sample estimate.



20. How are Z-scores used to compare different distributions?
*  Z-scores standardize values from different distributions by converting them into the same scale with a mean of 0 and a standard deviation of 1. This allows us to compare values across different datasets, even if the original units or spreads are different—for example, comparing test scores from two different exams.



21. What are the assumptions for applying the Central Limit Theorem?
*  To apply the Central Limit Theorem (CLT), we make a few key assumptions:
  1. Samples are independent and randomly selected.
  2. The sample size should be large enough
  3. If the population isn’t normal, a larger sample size is needed for the sample mean to be approximately normal.




22. What is the concept of expected value in a probability distribution?
*  The expected value is the long-term average or mean we expect from a probability distribution if we repeat the experiment many times. It’s calculated by multiplying each possible outcome by its probability and summing them up. It gives us a sense of the "center" of the distribution.



23. How does a probability distribution relate to the expected outcome of a random variable?
*  A probability distribution shows all possible values a random variable can take and their associated probabilities. The expected outcome (or expected value) is calculated from this distribution and represents the average result we’d expect over many trials. It connects the random variable to its most likely long-term behavior.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p


p = 0.7
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
# 18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
data = [50, 52, 47, 49, 53, 48, 51, 54, 46, 50]
x = 54

mean = np.mean(data)
std = np.std(data)

z_score = (x - mean) / std
print(f"Z-score for {x} is: {z_score:.2f}")

x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals, 0, 1)

plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z_score), color='red', alpha=0.3)

plt.title('Z-score Comparison with Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Sample data
sample = [52, 55, 53, 54, 56, 51, 50, 57, 55, 54]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)


pop_mean = 50


z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
print(f"Z-statistic: {z:.2f}")


p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")


alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

# Sample data
data = [52, 55, 53, 54, 56, 51, 50, 57, 55, 54]
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)


confidence = 0.95
z = norm.ppf(1 - (1 - confidence)/2)  # Z-score for 95% confidence ~1.96


margin_error = z * (std / np.sqrt(n))

lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"Sample Mean: {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy.stats import norm

np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, size=50

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence = 0.95
z = norm.ppf(1 - (1 - confidence) / 2)  # Z-score for 95% confidence (~1.96)

margin_error = z * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error


print(f"Sample Mean: {mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

# Generate data from a normal distribution
data = np.random.normal(loc=100, scale=15, size=1000)

# Calculate Range
data_range = np.max(data) - np.min(data)

# Calculate IQR (Q3 - Q1)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (normal distribution)
data = np.random.normal(loc=100, scale=15, size=1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Plot before and after normalization
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, edgecolor='black', color='orange')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # default is Fisher's definition (normal => 0)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f}")

# Generate normal data
np.random.seed(0)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Analyze
analyze_distribution(normal_data)
```
